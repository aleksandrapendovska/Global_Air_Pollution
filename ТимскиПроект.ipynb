#2. Предвидување на глобалното загадување на воздухот во различни земји

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/MyDrive/Тимски Проект /global air pollution dataset.csv')
```

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['AQI Category'] = le.fit_transform(df['AQI Category'])
```

```python
df['AQI Category'].value_counts()
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop(['Country', 'City', 'AQI Category', 'CO AQI Category', 'Ozone AQI Category', 'NO2 AQI Category', 'PM2.5 AQI Category'], axis = 1)
```

```python
x
```

```python
y = df.iloc[:, 3:4]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

classifier = DecisionTreeClassifier(max_depth=3)
classifier = classifier.fit(x_train,y_train)

y_pred = classifier.predict(x_test)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred))
```

```python
df_pred = pd.DataFrame(y_pred)
```

```python
df_pred.columns = ['AQI Category Predictions']
```

```python
df_pred
```

```python
category_mapping = {0: 'Good', 1: 'Hazardous', 2: 'Moderate', 3: 'Unhealthy', 4: 'Unhealthy for Sensitive Groups', 5: 'Very Unhealthy'}
```

```python
df_pred['AQI Category Predictions'] = df_pred['AQI Category Predictions'].map(category_mapping)
```

```python
df_pred
```

```python
df_pred['AQI Category Predictions'].value_counts().plot.bar()
```
